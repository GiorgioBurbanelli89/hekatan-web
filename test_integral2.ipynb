# Comparacion: handcalcs vs SymPy (init_printing) vs Hekatan

---
## 1. handcalcs (%%render)
Solo aritmetica basica. NO soporta integrales.

```python
import handcalcs.render
import math
```

```python
%%render
a = 3
b = 4
c = math.sqrt(a**2 + b**2)
```

---
## 2. SymPy con init_printing()
Renderiza automaticamente en LaTeX. Soporta integrales simbolicas.

```python
from sympy import *
init_printing()
```

```python
# Pitagoras simbolico
a, b = symbols('a b')
expr = sqrt(a**2 + b**2)
expr
```

```python
# Evaluar con valores
expr.subs([(a, 3), (b, 4)])
```

```python
# Integral (solo muestra, no evalua)
x = symbols('x')
Integral(3*x**2 + 2*x + 1, (x, 0, 2))
```

```python
# Integral evaluada
integrate(3*x**2 + 2*x + 1, (x, 0, 2))
```

---
## 3. Hekatan
HTML formateado. Variables, ecuaciones, formulas, integrales.

```python
from hekatan import *
import math
```

```python
# Pitagoras con Hekatan
var("a", 3)
var("b", 4)
formula("sqrt(a^2 + b^2)", "c")
c = math.sqrt(3**2 + 4**2)
eq("c", c)
```

```python
# Integral con nombre
integral("3*x^2 + 2*x + 1", "x", "0", "2", "I")
```

```python
# Integral sin nombre
integral("3*x^2 + 2*x + 1", "x", "0", "2")
```
