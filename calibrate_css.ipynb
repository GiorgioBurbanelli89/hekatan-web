# Calibración CSS: LaTeX (MathJax) vs Hekatan (HTML/CSS)
## Comparación visual lado a lado para ajustar el rendering

**Objetivo:** Comparar cómo renderiza LaTeX/MathJax cada expresión matemática versus cómo lo hace Hekatan con HTML/CSS puro, para calibrar que sean visualmente equivalentes.

````python
# === Imports ===
from IPython.display import display, Math, Latex, HTML
import hekatan as hk
from hekatan import calc, calc_clear
import math
````

````python
# Helper: muestra LaTeX y Hekatan lado a lado en una tabla HTML
def compare(title, latex_str, hekatan_fn):
    """Muestra título + LaTeX a la izquierda, Hekatan a la derecha."""
    display(HTML(f'<h4 style="margin:12px 0 4px;color:#444;border-bottom:1px solid #ddd;padding-bottom:4px">{title}</h4>'))
    display(HTML('<table style="width:100%;border-collapse:collapse"><tr>'
                 '<td style="width:50%;vertical-align:top;padding:8px;background:#fafafa;border:1px solid #eee">'
                 '<b style="font-size:10px;color:#888">LaTeX / MathJax</b></td>'
                 '<td style="width:50%;vertical-align:top;padding:8px;background:#f5f5ff;border:1px solid #eee">'
                 '<b style="font-size:10px;color:#888">Hekatan HTML/CSS</b></td></tr></table>'))
    # LaTeX
    display(Math(latex_str))
    # Hekatan
    hekatan_fn()
````

# 1. Fracciones

### LaTeX / MathJax

````python
# Fracción simple
display(Math(r'R = \frac{a + b}{c + d}'))
# Fracción con exponentes
display(Math(r'f(x) = \frac{x^2 + 1}{2x}'))
# Fracción ingeniería
display(Math(r'\sigma = \frac{F}{A} = \frac{25.5}{0.004} = 6375 \; \text{MPa}'))
# Fracción compuesta
display(Math(r'I = \frac{b \cdot h^3}{12} = \frac{300 \cdot 600^3}{12} = 5.4 \times 10^9 \; \text{mm}^4'))
````

### Hekatan HTML/CSS

````python
hk.fraction("a + b", "c + d", name="R")
hk.fraction("x^2 + 1", "2*x", name="f(x)")
hk.fraction("F", "A", name="sigma")
# Fracción con calc engine
calc_clear()
calc("b = 300", unit="mm")
calc("h = 600", unit="mm")
calc("I = b * h^3 / 12", unit="mm^4")
````

---
# 2. Integrales

### LaTeX / MathJax

````python
# Integral definida
display(Math(r'\int_0^L x^2 \, dx'))
# Integral trigonométrica
display(Math(r'\int_0^{\pi} \sin(x) \, dx = 2'))
# Integral gaussiana
display(Math(r'\int_{-\infty}^{\infty} e^{-x^2} \, dx = \sqrt{\pi}'))
# Integral doble
display(Math(r'\int_0^1 \int_0^1 x \cdot y \, dx \, dy'))
# Integral con resultado
display(Math(r'\int_0^L \frac{M(x)}{E \cdot I} \, dx = \frac{F \cdot L^3}{48 \cdot E \cdot I}'))
````

### Hekatan HTML/CSS

````python
hk.integral("x^2", variable="x", lower="0", upper="L")
hk.integral("sin(x)", variable="x", lower="0", upper="pi")
hk.integral("e^(-x^2)", variable="x", lower="-inf", upper="inf")
hk.double_integral("x*y", var1="x", var2="y", lower1="0", upper1="1", lower2="0", upper2="1")
hk.integral("M(x) / (E*I)", variable="x", lower="0", upper="L")
````

---
# 3. Derivadas

### LaTeX / MathJax

````python
# Primera derivada
display(Math(r'\frac{d}{dx}\left(x^3 + 2x\right) = 3x^2 + 2'))
# Segunda derivada
display(Math(r'\frac{d^2}{dx^2}\left(e^{x^2}\right)'))
# Derivada producto trig
display(Math(r'\frac{d}{dx}\left(\sin x \cdot \cos x\right) = \cos^2 x - \sin^2 x'))
# Derivada con notación prima
display(Math(r"y'' + \omega^2 y = 0"))
````

### Hekatan HTML/CSS

````python
hk.derivative("x^3 + 2*x", variable="x")
hk.derivative("e^(x^2)", variable="x", order=2)
hk.derivative("sin(x)*cos(x)", variable="x")
hk.formula("y'' + omega^2 * y = 0", name="")
````

# 4. Derivadas Parciales

### LaTeX / MathJax

````python
display(Math(r'\frac{\partial}{\partial x}\left(x^2 y + y^3\right) = 2xy'))
display(Math(r'\frac{\partial}{\partial y}\left(x^2 y + y^3\right) = x^2 + 3y^2'))
# Laplaciano
display(Math(r'\nabla^2 u = \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0'))
# Derivada parcial mixta
display(Math(r'\frac{\partial^2 f}{\partial x \, \partial y}'))
````

### Hekatan HTML/CSS

````python
hk.partial("x^2*y + y^3", variable="x")
hk.partial("x^2*y + y^3", variable="y")
hk.partial("u", variable="x", order=2)
hk.partial("u", variable="y", order=2)
````

# 5. Sumatorias

### LaTeX / MathJax

````python
display(Math(r'\sum_{k=1}^{n} k^2 = \frac{n(n+1)(2n+1)}{6}'))
display(Math(r'\sum_{k=1}^{\infty} \frac{1}{k}'))
# Serie de Taylor
display(Math(r'e^x = \sum_{n=0}^{\infty} \frac{x^n}{n!}'))
# Suma de fuerzas
display(Math(r'\sum_{i=1}^{n} F_i = 0'))
````

````python
hk.summation("k^2", variable="k", lower="1", upper="n")
hk.summation("1/k", variable="k", lower="1", upper="inf")
hk.summation("x^n / n!", variable="n", lower="0", upper="inf")
hk.summation("F_i", variable="i", lower="1", upper="n")
````

# 7. Límites

### LaTeX / MathJax

````python
display(Math(r'\lim_{x \to 0} \frac{\sin x}{x} = 1'))
display(Math(r'\lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n = e'))
display(Math(r'\lim_{h \to 0} \frac{f(x+h) - f(x)}{h}'))
display(Math(r'\lim_{x \to 0^+} \frac{1}{x} = +\infty'))
````

# 10. Ecuaciones Diferenciales

### LaTeX / MathJax

````python
# Oscilador armónico
display(Math(r"y'' + \omega^2 y = 0"))
display(Math(r'y(x) = C_1 \sin(\omega x) + C_2 \cos(\omega x)'))
# Ecuación de calor
display(Math(r'\frac{\partial u}{\partial t} = \alpha \frac{\partial^2 u}{\partial x^2}'))
# Ecuación de viga
display(Math(r'EI \frac{d^4 y}{dx^4} = q(x)'))
# Ecuación de onda
display(Math(r'\frac{\partial^2 u}{\partial t^2} = c^2 \frac{\partial^2 u}{\partial x^2}'))
# Laplaciana
display(Math(r'\nabla^2 \phi = \frac{\partial^2 \phi}{\partial x^2} + \frac{\partial^2 \phi}{\partial y^2} + \frac{\partial^2 \phi}{\partial z^2} = 0'))
````

### Hekatan HTML/CSS

# 11. Ecuaciones de Ingeniería Completas

### LaTeX / MathJax

````python
# Momento de inercia
display(Math(r'I = \frac{b \cdot h^3}{12} = \frac{300 \cdot 600^3}{12} = 5.4 \times 10^9 \; \text{mm}^4'))
# Esfuerzo flexión
display(Math(r'\sigma = \frac{M \cdot c}{I} = \frac{150 \times 10^3 \cdot 300}{5.4 \times 10^9} = 8.33 \; \text{MPa}'))
# Deflexión
display(Math(r'\delta_{\max} = \frac{5 q L^4}{384 E I}'))
# Fórmula cuadrática
display(Math(r'x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}'))
# Euler buckling
display(Math(r'P_{cr} = \frac{\pi^2 E I}{(K L)^2}'))
````

# Notas de Calibración CSS

## Aspectos a comparar visualmente:

| Elemento | LaTeX/MathJax | Hekatan CSS | Notas |
|----------|---------------|-------------|-------|
| Fracciones | `\frac{}{}` | `.dvc > .dvl` | Alineación vertical, grosor de línea |
| Integrales | `\int_{}^{}` | `.nary-wrap > .nary` | Tamaño símbolo, posición límites |
| Derivadas | `\frac{d}{dx}` | `.dvc > .dvl` | Espaciado d/dx |
| Parciales | `\frac{\partial}{\partial x}` | `.dvc > .dvl` | Símbolo ∂ |
| Sumatorias | `\sum_{}^{}` | `.nary-wrap > .nary` | Tamaño Σ, alineación |
| Productorias | `\prod_{}^{}` | `.nary-wrap > .nary` | Tamaño ∏ |
| Límites | `\lim_{x\to}` | custom spans | "lim" italic, flecha |
| Raíces | `\sqrt{}` | `.r + .o0` | Largo overline, símbolo √ |
| Matrices | `\begin{bmatrix}` | `table.matrix` | Brackets, alineación celdas |
| Superíndices | `^{}` | `<sup>` | Tamaño, posición |
| Subíndices | `_{}` | `<sub>` | Tamaño, posición |
| Font | MJXTEX serif | MJXTEX/STIX serif | Consistencia tipográfica |

## Fuentes CSS de Hekatan:
```css
font-family: 'MJXTEX', 'STIX Two Math', 'Cambria Math', 'Latin Modern Math', 'Times New Roman', serif;
```
