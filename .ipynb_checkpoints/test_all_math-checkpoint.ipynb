# Comparación Completa: handcalcs vs SymPy vs Hekatan
## Todas las formas de expresiones matemáticas en Jupyter

```python
# === Imports para las 3 librerías ===
import handcalcs.render
import sympy as sp
from sympy import symbols, sqrt as sp_sqrt, sin, cos, tan, exp, log, pi, oo, Integral, Derivative, Sum, Product, Limit, Matrix, Rational, Function, latex
sp.init_printing(use_latex='mathjax')

from hekatan import *
import math
```

---
# 1. Operaciones Básicas (sqrt, potencias)

### handcalcs

```python
%%render
a = 3
b = 4
c = math.sqrt(a**2 + b**2)
```

### SymPy

```python
a_s, b_s = symbols('a b')
expr = sp_sqrt(a_s**2 + b_s**2)
display(sp.Eq(sp.Symbol('c'), expr))
display(sp.Eq(sp.Symbol('c'), expr.subs({a_s: 3, b_s: 4})))
```

### Hekatan

```python
%%hekatan
a = 3
b = 4
c = math.sqrt(a**2 + b**2)
```

---
# 2. Fracciones / División

### handcalcs

```python
%%render
F = 25.5
A_s = 0.004
sigma = F / A_s
```

### SymPy

```python
F_s, A_s_sym = symbols('F A_s')
sigma_s = sp.Symbol('sigma')
display(sp.Eq(sigma_s, F_s / A_s_sym))
display(sp.Eq(sigma_s, (F_s / A_s_sym).subs({F_s: 25.5, A_s_sym: 0.004})))
```

### Hekatan (%%hekatan)

```python
%%hekatan
F = 25.5
A_s = 0.004
sigma = F / A_s
```

### Hekatan (fraction API)

```python
fraction("a + b", "c + d", name="R")
fraction("x^2 + 1", "2*x", name="f(x)")
fraction("F", "A", name="sigma", unit="MPa")
```

---
# 3. Integrales

### SymPy

```python
x = symbols('x')
L = sp.Symbol('L')

# Integral definida
display(Integral(x**2, (x, 0, L)))
# Integral con sin
display(Integral(sin(x), (x, 0, pi)))
# Gaussiana
display(Integral(exp(-x**2), (x, -oo, oo)))
# Evaluada
display(sp.Eq(Integral(sin(x), (x, 0, pi)), Integral(sin(x), (x, 0, pi)).doit()))
```

### Hekatan

```python
integral("x^2", var="x", lower="0", upper="L")
integral("sin(x)", var="x", lower="0", upper="pi")
integral("e^(-x^2)", var="x", lower="-inf", upper="inf")
```

### Hekatan — Integral doble

```python
double_integral("x*y", var1="x", var2="y", lower1="0", upper1="1", lower2="0", upper2="1")
```

---
# 4. Derivadas

### SymPy

```python
x = symbols('x')

# Primera derivada
display(Derivative(x**3 + 2*x, x))
display(sp.Eq(Derivative(x**3 + 2*x, x), Derivative(x**3 + 2*x, x).doit()))

# Segunda derivada
display(Derivative(exp(x**2), x, 2))

# Producto de trig
display(sp.Eq(Derivative(sin(x)*cos(x), x), Derivative(sin(x)*cos(x), x).doit()))
```

### Hekatan

```python
derivative("x^3 + 2*x", var="x")
derivative("sin(x)*cos(x)", var="x")
derivative("e^(x^2)", var="x", order=2)
```

---
# 5. Derivadas Parciales

### SymPy

```python
x, y = symbols('x y')
f = x**2 * y + y**3

display(sp.Eq(Derivative(f, x), Derivative(f, x).doit()))
display(sp.Eq(Derivative(f, y), Derivative(f, y).doit()))
```

### Hekatan

```python
partial("x^2*y + y^3", var="x")
partial("x^2*y + y^3", var="y")
```

---
# 6. Sumatorias

### SymPy

```python
k, n = symbols('k n')

display(Sum(k**2, (k, 1, n)))
display(sp.Eq(Sum(k**2, (k, 1, n)), Sum(k**2, (k, 1, n)).doit()))
display(Sum(1/k, (k, 1, oo)))
```

### Hekatan

```python
summation("k^2", var="k", lower="1", upper="n")
summation("1/k", var="k", lower="1", upper="inf")
```

---
# 7. Productorias

### SymPy

```python
k, n = symbols('k n')

display(Product(k, (k, 1, n)))
display(Product(1 + 1/k, (k, 1, n)))
```

### Hekatan

```python
product_op("k", var="k", lower="1", upper="n")
product_op("(1 + 1/k)", var="k", lower="1", upper="n")
```

---
# 8. Límites

### SymPy

```python
x, n = symbols('x n')

display(Limit(sin(x)/x, x, 0))
display(sp.Eq(Limit(sin(x)/x, x, 0), Limit(sin(x)/x, x, 0).doit()))

display(Limit((1 + 1/n)**n, n, oo))
display(sp.Eq(Limit((1 + 1/n)**n, n, oo), Limit((1 + 1/n)**n, n, oo).doit()))
```

### Hekatan

```python
limit_op("sin(x)/x", var="x", to="0")
limit_op("(1 + 1/n)^n", var="n", to="inf")
```

---
# 9. Matrices

### SymPy

```python
A_mat = Matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
display(sp.Eq(sp.MatrixSymbol('A', 3, 3), A_mat, evaluate=False))

K_mat = Matrix([[12, 6, -12, 6], [6, 4, -6, 2], [-12, -6, 12, -6], [6, 2, -6, 4]])
display(sp.Eq(sp.MatrixSymbol('K', 4, 4), K_mat, evaluate=False))
```

### Hekatan

```python
A_data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
matrix(A_data, "A")

K_data = [[12, 6, -12, 6], [6, 4, -6, 2], [-12, -6, 12, -6], [6, 2, -6, 4]]
matrix(K_data, "K")
```

# 10. Ecuaciones Diferenciales

### SymPy — EDO simbólica

```python
x = symbols('x')
y_fn = Function('y')

# y'' + y = 0 (oscilador armónico)
ode = sp.Eq(y_fn(x).diff(x, 2) + y_fn(x), 0)
display(ode)
sol = sp.dsolve(ode, y_fn(x))
display(sol)

# y' = -2*x*y (Gaussiana)
ode2 = sp.Eq(y_fn(x).diff(x), -2*x*y_fn(x))
display(ode2)
display(sp.dsolve(ode2, y_fn(x)))
```

### Hekatan — Ec. diferencial de viga

```python
# Ecuación diferencial de la elástica
formula("E*I*d^2y/dx^2 = M(x)", name="")
derivative("y(x)", var="x", order=2)
formula("y'' + omega^2 * y = 0", name="")
```

# 11. Trigonometría

### handcalcs

```python
%%render
theta = math.pi / 4
fx = math.sin(theta)
fy = math.cos(theta)
t = math.tan(theta)
```

### SymPy

```python
theta = symbols('theta')
display(sp.Eq(sp.Symbol('f_x'), sin(theta)))
display(sp.Eq(sp.Symbol('f_y'), cos(theta)))
display(sp.Eq(sp.Symbol('t'), tan(theta)))

# Identidad
display(sp.Eq(sin(theta)**2 + cos(theta)**2, 1))
```

### Hekatan

```python
%%hekatan
theta = math.pi / 4
fx = math.sin(theta)
fy = math.cos(theta)
t = math.tan(theta)
```

# 13. Raíz Cuadrada

### SymPy

```python
a_s, b_s, c_s = symbols('a b c')
display(sp.Eq(c_s, sp_sqrt(a_s**2 + b_s**2)))

x_s, y_s, z_s, r_s = symbols('x y z r')
display(sp.Eq(r_s, sp_sqrt(x_s**2 + y_s**2 + z_s**2)))
```

### Hekatan

```python
sqrt("a^2 + b^2", name="c")
sqrt("x^2 + y^2 + z^2", name="r")
```

# 15. Calc Engine — Evaluación con pasos (Solo Hekatan)

```python
calc_clear()
calc("b = 300", unit="mm")
calc("h = 600", unit="mm")
calc("A = b * h", unit="mm^2")
calc("I = b * h^3 / 12", unit="mm^4")
```

```python
calc_clear()
calc("F = 25.5", unit="kN")
calc("L = 6.0", unit="m")
calc("M = F * L / 4", unit="kN*m")
calc("V = F / 2", unit="kN")
```

# 16. eq_block — Ecuación completa con sustitución (Solo Hekatan)

# Resumen Comparativo

| Expresión | handcalcs | SymPy | Hekatan |
|---|---|---|---|
| Asignaciones Python | ✅ %%render | ❌ | ✅ %%hekatan |
| Fracciones | auto | ✅ Rational | ✅ fraction() |
| Integrales | ❌ | ✅ Integral | ✅ integral() |
| Derivadas | ❌ | ✅ Derivative | ✅ derivative() |
| Derivadas parciales | ❌ | ✅ Derivative | ✅ partial() |
| Sumatorias | ❌ | ✅ Sum | ✅ summation() |
| Productorias | ❌ | ✅ Product | ✅ product_op() |
| Límites | ❌ | ✅ Limit | ✅ limit_op() |
| Raíz cuadrada | auto | ✅ sqrt | ✅ sqrt() |
| Matrices | ❌ | ✅ Matrix | ✅ matrix() |
| Ec. Diferenciales | ❌ | ✅ dsolve | ✅ formula() |
| Unidades | ❌ | ❌ | ✅ eq(), var() |
| Verificaciones | ❌ | ❌ | ✅ check() |
| Calc engine | ❌ | ❌ | ✅ calc() |
| eq_block (3 pasos) | ❌ | ❌ | ✅ eq_block() |
| Layout (títulos) | ❌ | ❌ | ✅ title(), text() |
| Motor rendering | LaTeX/MathJax | LaTeX/MathJax | HTML/CSS |
| Dependencias | IPython | sympy | 0 (stdlib) |
