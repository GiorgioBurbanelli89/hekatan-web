# %%render (handcalcs) vs %%hekatan
Mismo codigo Python, diferente output

---
## 1. handcalcs (%%render) — LaTeX/MathJax

```python
import handcalcs.render
import math
```

```python
%%render
a = 3
b = 4
c = math.sqrt(a**2 + b**2)
```

---
## 2. Hekatan (%%hekatan) — HTML/CSS

```python
from hekatan import *
import math
```

```python
%%hekatan
a = 3
b = 4
c = math.sqrt(a**2 + b**2)
```

---
## 3. Hekatan con mas formulas

```python
%%hekatan
F = 25.5
A = 0.004
sigma = F / A
```
